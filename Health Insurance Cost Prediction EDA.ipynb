```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

# Handling Missing values

```python
df.isnull().sum()
```

```python
# No missing Values
```

# AGE VS CHARGES

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x = "age", y = "charges", data=df)
sns.regplot(x = "age", y = "charges", data=df, scatter=False, color = 'red')
plt.title("Age VS Charges")
```

# BMI VS CHARGES

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x = "bmi", y = "charges", data=df)
sns.regplot(x = "bmi", y = "charges", data=df, scatter=False, color = 'red')
plt.title("BMI VS Charges")
```

# Smoker Status vs Charges

```python
plt.figure(figsize=(8,5))
sns.boxenplot(x = "smoker", y = "charges", data=df)
plt.title("Smoker status VS Charges")
```

# Region vs Charges

```python
plt.figure(figsize=(8,5))
sns.boxenplot(x = "region", y = "charges", data=df)
plt.title("Region VS Charges")
```
